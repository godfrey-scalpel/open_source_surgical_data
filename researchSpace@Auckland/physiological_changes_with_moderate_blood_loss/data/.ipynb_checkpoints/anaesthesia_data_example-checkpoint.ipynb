```python
import pandas as pd
import xml.etree.ElementTree as et
```

```python
root_patient17 = et.parse("17_2.xml").getroot()
```

```python
print(et.tostring(root_patient17, encoding='utf8').decode('utf8'))
```

```python
for child in root_patient17:
    print(child.tag, child.text)
    for grandchild in child:
        print(grandchild.tag)
```

```python
events = [y for y in root_patient17 if y.tag == 'events']
```

```python
for event in events:
    print(event)
```
